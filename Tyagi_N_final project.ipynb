# ANOVA between duration and IMDB_score

```python
df['duration'].describe()
```

```python
df2 = df
# setting values for bins
df2['duration_bins'] = pd.cut(df['duration'], bins=[30, 100, 170, 250, 330], labels=[1,2,3,4])
# see the result
df2.head()

#'too short','short','medium','long'= 1,2,3,4
```

```python
df2.boxplot('imdb_score', by='duration_bins', figsize=(12, 8));
```

## Null Hypothesis : the duration of the movie does not affect IMDB score

```python
aov = pg.anova(data=df2, dv='imdb_score', between='duration_bins', detailed=True)
aov
```

```python
pg.pairwise_tukey(data=df2, dv='imdb_score', between='duration_bins')
import scipy.stats as stats

print(stats.normaltest(df2['duration_bins']))
```

```python
#'too short','short','medium','long'= 1,2,3,4

db_1 = df2[df2['duration_bins'] == 1]['imdb_score']
db_2 = df2[df2['duration_bins'] == 2]['imdb_score']
db_3 = df2[df2['duration_bins'] == 3]['imdb_score']
db_4 = df2[df2['duration_bins'] == 4]['imdb_score']
f_val, p_val = stats.f_oneway(db_1, db_2, db_3, db_4)  
  
print("One-way ANOVA P =", p_val)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

mod = ols('imdb_score ~ duration_bins', data=df2).fit()
                
aov_table = sm.stats.anova_lm(mod)
print(aov_table)

```

```python
# post hoc testing

from statsmodels.stats.multicomp import pairwise_tukeyhsd

print(pairwise_tukeyhsd(df2['imdb_score'], df2['duration_bins']))
```

**The Analysis of Variance shows significant differences between duration and IMDB_score. The p-value is less than 0.05.
According to the key ('too short','short','medium','long'= 1,2,3,4). The shorter movies are have significant impact on the score.**

```python
from sklearn.cluster import KMeans

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances
```

```python
#The Elbow method

#https://www.packtpub.com/big-data-and-business-intelligence/mastering-machine-learning-scikit-learn#

#http://docs.scipy.org/doc/scipy/reference/spatial.distance.html
#Computes distance between each pair of the two collections of inputs
from scipy.spatial.distance import cdist 

K = range(1, 10) 

meandistortions = []

for k in K: 
    kmeans = KMeans(n_clusters=k, random_state=1) 
    kmeans.fit(X) 
    meandistortions.append(sum(np.min(cdist(X, kmeans.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0]) 

plt.plot(K, meandistortions, 'bx-') 
plt.xlabel('k') 
plt.ylabel('Average distortion') 
plt.title('Selecting k with the Elbow Method')

#using scikit-plot python package (https://github.com/reiinakano/scikit-plot) ... you need to install 
# !pip install scikit-plot

import scikitplot as skplt

kmeans = KMeans(random_state=1) 
skplt.cluster.plot_elbow_curve(kmeans, X, cluster_ranges=range(1, 10))
```
